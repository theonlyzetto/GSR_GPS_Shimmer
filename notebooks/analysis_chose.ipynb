from pathlib import Path
import ipywidgets as widgets
from IPython.display import display, clear_output

data_dir = (project_root / "data" / "raw").resolve()
print("Datendirectory:", data_dir)

# Alle CSVs & DBs in data/raw finden
gsr_files = sorted(data_dir.glob("*.csv"))
db_files = sorted(data_dir.glob("*.db"))

print("Gefundene CSVs:")
for f in gsr_files:
    print("  -", f.name)

print("Gefundene DBs:")
for f in db_files:
    print("  -", f.name)

if not gsr_files or not db_files:
    print("‚ö†Ô∏è Achtung: Es wurden nicht sowohl CSV als auch DB gefunden!")


# Dropdowns f√ºr GSR-CSV & DB
gsr_dropdown = widgets.Dropdown(
    options=[(f.name, str(f)) for f in gsr_files],
    description="GSR-CSV:",
    layout=widgets.Layout(width="500px"),
    style={'description_width': 'initial'}
)

db_dropdown = widgets.Dropdown(
    options=[(f.name, str(f)) for f in db_files],
    description="DB-Datei:",
    layout=widgets.Layout(width="500px"),
    style={'description_width': 'initial'}
)

run_button = widgets.Button(
    description="Pipeline starten",
    button_style="success",
    icon="play"
)

output_box = widgets.Output()

def on_run_clicked(b):
    with output_box:
        clear_output()
        print("Starte Pipeline mit:")
        print("  GSR:", gsr_dropdown.label if hasattr(gsr_dropdown, 'label') else gsr_dropdown.value)
        print("  DB :", db_dropdown.label if hasattr(db_dropdown, 'label') else db_dropdown.value)
        res = run_gsr_gps_pipeline(
            gsr_csv_path=gsr_dropdown.value,
            db_path=db_dropdown.value,
            out_dir=str((project_root / "results").resolve())
        )
        print("\n‚úÖ Pipeline fertig. Ergebnisse:")
        for k, v in res.items():
            print(k, "‚Üí", v)

run_button.on_click(on_run_clicked)

display(widgets.VBox([
    widgets.HTML("<b>Bitte CSV & DB w√§hlen und dann Pipeline starten:</b>"),
    gsr_dropdown,
    db_dropdown,
    run_button,
    output_box
]))
